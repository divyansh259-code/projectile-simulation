```python
import matplotlib.pyplot as plt
from project import trajectory

time, x_id, y_id, x_drag, y_drag = trajectory(v0=45, angle_degrees=45, drag_coeff=0.15)
plt.figure(figsize=(9, 5))
plt.plot(x_id, y_id, label='Vacuum (Ideal)', color='black', linestyle='--')
plt.plot(x_drag, y_drag, label='With Air Resistance', color='teal', linewidth=2)

plt.title('Projectile Motion Simulation Analysis', fontsize=12, fontweight='bold')
plt.xlabel('Distance (m)')
plt.ylabel('Height (m)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
